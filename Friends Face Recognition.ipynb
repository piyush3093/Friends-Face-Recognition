```python
#importing the necessery libraries
import cv2
import face_recognition
import matplotlib.pyplot as plt
from scipy.spatial import distance
```

```python
#importing the images
chandler = cv2.imread('Images/chandler.jpg')
joey = cv2.imread('Images/joey.jpg')
monica = cv2.imread('Images/monica.jpg')
phoebe = cv2.imread('Images/phoebe.jpg')
rachel = cv2.imread('Images/rachel.jpg')
ross = cv2.imread('Images/ross.jpg')
```

```python
chandler = cv2.cvtColor(chandler, cv2.COLOR_BGR2RGB)
joey = cv2.cvtColor(joey, cv2.COLOR_BGR2RGB)
monica = cv2.cvtColor(monica, cv2.COLOR_BGR2RGB)
phoebe = cv2.cvtColor(phoebe, cv2.COLOR_BGR2RGB)
rachel = cv2.cvtColor(rachel, cv2.COLOR_BGR2RGB)
ross = cv2.cvtColor(ross, cv2.COLOR_BGR2RGB)
```

```python
images = []
images.append(chandler)
images.append(joey)
images.append(monica)
images.append(phoebe)
images.append(rachel)
images.append(ross)
```

```python
print(len(images))
```

```python
characters = ['Chandler', 'Joey', 'Monica', 'Phoebe', 'Rachel', 'Ross', 'Other']
```

```python
fig = plt.figure(figsize = (15, 4))
for i in range(0, len(images)):
    ax = fig.add_subplot(2, 3, i+1, xticks = [], yticks = [])
    plt.imshow(images[i])
    ax.set_title(characters[i])
```

```python
#encoding the faces
encoded_images = []
for i in range(0, len(images)):
    encoded_images.append(face_recognition.face_encodings(images[i]))
```

```python
#importing the test images
test_images = []
test_images.append(cv2.imread('Test/chandler_and_monica.jpg'))
test_images.append(cv2.imread('Test/monica_rachel_phoebe.jpg'))
test_images.append(cv2.imread('Test/phoebe_test.png'))
test_images.append(cv2.imread('Test/ross_test.jpg'))
test_images.append(cv2.imread('Test/test_everyone.jpg'))
test_images.append(cv2.imread('Test/test_everyone_2.jpeg'))
```

```python
for i in range(0, len(test_images)):
    test_images[i] = cv2.cvtColor(test_images[i], cv2.COLOR_BGR2RGB)
```

```python
fig = plt.figure(figsize = (15, 4))
for i in range(0, len(test_images)):
    ax = fig.add_subplot(2, 3, i+1, xticks = [], yticks = [])
    plt.imshow(test_images[i])
```

```python
#encoding the test faces
test_faces = []
for image in test_images:
    faces_encode = face_recognition.face_encodings(image)
    test_faces.append(faces_encode)
```

```python
print(type(test_faces))
print(len(test_faces))
```

```python
test_characters = []
for i in range(0, len(test_faces)):
    image_characters = []
    for j in range(0, len(test_faces[i])):
        flag = False
        for k in range(0, len(encoded_images)):
            d = distance.euclidean(test_faces[i][j], encoded_images[k])
            if(d < 0.6):
                flag  = True
                image_characters.append(characters[k])
        if(flag == False):
            image_characters.append(characters[6])
    test_characters.append(image_characters)
```

```python
print(test_characters[1])
```

```python
#Visualizing the Results
face_locations = face_recognition.face_locations(test_images[0])
plt.imshow(test_images[0])
ax = plt.gca()
for i in range(len(face_locations)):
    top, right, bottom, left = face_locations[i]
    x, y, w, h = left, top, right, bottom
    rect = plt.Rectangle((x, y), w-x, h-y, fill = False, color = 'green')
    ax.add_patch(rect)
    ax.text(x, h, test_characters[0][i], color = 'red', fontsize = 'large', fontweight = 'heavy')
plt.show()
```
